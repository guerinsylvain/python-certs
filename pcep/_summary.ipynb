```python
# Extra: Watch out for integer division!
# In the video about Operators, I mention that integer division (//) always returns the nearest whole number rounded down.
# Keep in mind, however, that this whole number might be of type integer or float, depending on your input numbers:

6 // 3     # returns 2
6.0 // 3   # returns 2.0
6 // 3.0   # returns 2.0
6.0 // 3.0 # returns 2.0
If you integer divide two integers, the result is of type integer. But if at least one of the input numbers is a float, 
you will get a float instead. The same rule applies to all other operators except for the standard division (/),
which always returns a float, even for two integers.

This can be a bit tricky so keep that in mind for the exam :)
```

```python
# functions
def greet():
    print('Hello, my dear!')
    
greet()
greet()

# default values
def print_letter_count(text='This is the default string to search', letter='a'):
    counter = 0
    for char in text:
        if char == letter:
            counter += 1
    print('Number of', letter, 'is', counter)
    
def show_truth():
    # This is called shadowing !!!
    mysterious_var = 'New Surprise!'
    print(mysterious_var)

mysterious_var = 'Surprise!'
print(mysterious_var)
show_truth()
print(mysterious_var)

def show_truth():
    # To prevent shadowing !!!
    global mysterious_var
    mysterious_var = 'New Surprise!'
    print(mysterious_var)

mysterious_var = 'Surprise!'
print(mysterious_var)
show_truth()
print(mysterious_var)

def show_truth():
    # No shadowing here because no variable assignment !!!
    # This rule applies to Lists and dictionaries becaure they are mutable
    # This rule does not apply to tupple because tupples are immutable
    mysterious_var.append('New Surprise!')
    print(mysterious_var)

mysterious_var = ['Surprise!']
print(mysterious_var)
show_truth()
print(mysterious_var)
```

```python
# Generators

def get_number():
    for i in range(1, 4):
        yield i

print(get_number())        

for x in get_number():
    print(x)
    
numbers = list(get_number())
print(numbers)    
```
